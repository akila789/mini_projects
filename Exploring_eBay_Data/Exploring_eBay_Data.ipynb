## Analysing Used cars in eBay Germany

In this project we analyse dataset of used cars from *eBay Kleinanzeigen*, a used cars section of the Germany eBay website.

The dataset originally scraped and uploaded to [Kaggle](http://www.kaggle.com). 

Our aim of this project is to clean the dataset and analyze included used car listings.

Let's start by importing libraries we need to import the dataset into pandas.

The data dictionary provided with data is as follows:

- dateCrawled - When this ad was first crawled. All field-values are taken from this date.
- name - Name of the car.
-  seller - Whether the seller is private or a dealer.
- offerType - The type of listing
- price - The price on the ad to sell the car.
- abtest - Whether the listing is included in an A/B test.
- vehicleType - The vehicle Type.
- yearOfRegistration - The year in which the car was first registered.
- gearbox - The transmission type.
- powerPS - The power of the car in PS.
- model - The car model name.
- kilometer - How many kilometers the car has driven.
- monthOfRegistration - The month in which the car was first registered.
- fuelType - What type of fuel the car uses.
- brand - The brand of the car.
- notRepairedDamage - If the car has a damage which is not yet repaired.
- dateCreated - The date on which the eBay listing was created.
- nrOfPictures - The number of pictures in the ad.
- postalCode - The postal code for the location of the vehicle.
- lastSeenOnline - When the crawler saw this ad last online.

####  Exploring price by brand

```python
autos['brand'].value_counts(normalize = True)
```

German manufacturers represent top 5 brands, almost 61% of total vehicles. Volkswagen is the dominant brand which contain 21% from total dataset.

Also there are lot of brands not having significant share from the dataset. Therefore we limit our analysis for vehicle brands which having atleast 5% from total.

```python
brand_count = autos['brand'].value_counts(normalize = True)
common_brands = brand_count[brand_count > 0.05].index
print(common_brands)
```

```python
brand_mean_prices = {}
for brand in common_brands:
    prices = autos[autos['brand'] == brand]["price"].mean()
    brand_mean_prices[brand] = int(prices)
brand_mean_prices_series = pd.Series(brand_mean_prices)
brand_data = pd.DataFrame(brand_mean_prices_series, columns=['mean_price'])
brand_data
```

From the top 6 brands Audi ,Mercedes Benz and BMW are more expensive. Ford and Opel are less expensive. Volkswagen in between. This can explain its popularity.

#### Exploring Mileage

```python
mean_mileage = {}
for brand in common_brands:
    mileage = autos[autos['brand']== brand]['odometer_km'].mean()
    mean_mileage[brand] = mileage
mean_milaege_series = pd.Series(mean_mileage)
```

```python
mean_mileage
```

```python
brand_data['mean_mileage'] = mean_milaege_series
```

```python
brand_data
```

The range of car milaege not vary by brand. But there is slight trend to the more expensive vehicles having higher mileage and less expensive vehicle having low mileage.
